```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

pd.pandas.set_option("display.max_columns", None)
```

```python
data = pd.read_csv("train.csv")
data.head()
```

# Missing Values

```python
## First lets handle Categorical features which are missing

features_with_na = [i for i in data.columns if data[i].isnull().any() and data[i].dtype == 'O']

for i in features_with_na:
    print(i, "-> ", np.round(data[i].isnull().mean(), 4), "% missing")
```

```python
# Generic function to replace vareiables with a value

def replace_vals(df, features, replace_val):
    new_df = df.copy()
    new_df[features] = new_df[features].fillna(replace_val)
    return new_df

data = replace_vals(data, features_with_na, "Missing")

data[features_with_na].isnull().sum()
```

```python
# numerical values containing the missing values

num_features_na = [i for i in data.columns if data[i].isnull().any() and data[i].dtype != 'O']

for i in num_features_na:
    print(i, " -> ", np.round(data[i].isnull().mean(), 4),"% Missing values")
```

```python
# Replacing nan values and keeping track of missing values 
for i in num_features_na:
    
    median_val = data[i].median()
    data[i+"_nan"] = np.where(data[i].isnull(), 1, 0)
    
    data[i].fillna(median_val, inplace=True)
    
# to verify the replaced nan values    
data[num_features_na].isnull().sum()
```

```python
data.head(20)
```

```python
# Temporal variables (date time variables)- trying to find the relationship between Yrsold and other temporal variables

for feature in ["GarageYrBlt","YearRemodAdd","YearBuilt"]:
    
    data[feature] = data["YrSold"] - data[feature]
```

```python
data[["GarageYrBlt","YearRemodAdd","YearBuilt"]].head()
```

```python
data.head()
```

### Since the numerical variables are skewed we will perform the log normal distribution on them to so that it will follow the gaussian distribution 

```python

num_features = ["LotFrontage", "LotArea", "GrLivArea", "1stFlrSF" ,"SalePrice"]

for feature in num_features_na:
    data[feature] = np.log(data[feature])

data.head()
```

```python
categorical_features = [i for i in data.columns if data[i].dtype =='O']

categorical_features
```

```python
for feat in categorical_features:
    
    temp = data.groupby(feat)['SalePrice'].count()/len(data)
    temp_df = temp[temp>0.01].index
    data[feat] = np.where(data[feat].isin(temp_df), data[feat], 'Rare_var')
```

```python
for feat in categorical_features:
    
    labels = data.groupby(feat)['SalePrice'].mean().index
    labels_mapping = {k:v for v,k in enumerate(labels,0)}
    data[feat] = data[feat].map(labels_mapping)
```

```python
data.head()
```

# Feature Scaling

```python
scaling_feat = [feat for feat in data.columns if feat not in ['Id', 'SalePrice']]
scaling_feat
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[scaling_feat])
```

```python
scaler.transform(data[scaling_feat])
```

```python
data = pd.concat([data[["Id","SalePrice"]].reset_index(drop=True),
         pd.DataFrame(scaler.transform(data[scaling_feat]), columns=scaling_feat)],
         axis=1)
```
